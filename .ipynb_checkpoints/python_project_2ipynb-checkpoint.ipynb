Merging all the tables



```python

orders.rename(columns = {'order_id':'id_order'}, inplace = True)
orders
products_sales = orders.merge(orderlines, how= 'inner', on = 'id_order').merge(products, how = 'inner', on ='sku')
products_sales['created_date'] =pd.to_datetime(products_sales['created_date'])
products_sales['date'] =pd.to_datetime(products_sales['date'])

products_sales.info()
```

```python
products_sales_c = products_sales.query('state == "Completed"')
products_total_sales= (
products_sales_c
    #.assign(year_month =products_sales_c['date'].dt.strftime('%Y-%m'))
    .groupby(['category'])
    .agg({'total_price':'sum'}) 
    .reset_index()
    .sort_values('total_price', ascending=False)
)
products_total_sales

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
colors = sns.color_palette('deep')
plt.pie(products_total_sales['total_price'], autopct='%1.1f%%', startangle=90, radius= 3 , colors=colors)
plt.legend(products_total_sales['category'], loc="upper right" )


#plt.title('Distribution of Product Categories', fontsize = 17)
plt.show()
```

```python
p_imac= products_sales_c.query('category == "iMac"')
p_imac_revenue = ( p_imac
   
    .assign(year_month = p_imac['date'].dt.strftime('%b-%Y'))
    .assign(year = p_imac['date'].dt.strftime('%Y'))
    .assign(month = p_imac['date'].dt.strftime('%m'))
    .groupby([ 'year_month', 'year', 'month'])
    .agg({'total_price':'sum'})
    .reset_index()
    .sort_values( ["year", "month"])

)
p_imac_revenue
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.subplots(figsize=(14, 6))
sns.lineplot(data=p_imac_revenue, x='year_month', y='total_price', color ='red')
plt.title('Total Revenue generated by the iMac from Jan 2017-Mar 2018', fontsize = 17)
plt.xlabel("Timeline (months)", fontsize = 12)
plt.ylabel("Total Revenue", fontsize = 12);
```
